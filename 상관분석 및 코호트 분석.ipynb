```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

from datetime import datetime 
from sklearn.datasets import load_iris
```

### 상관관계 시각화

```python
iris = load_iris()
iris
```

```python
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df.head(3)
#tail 
```

```python
target_df = pd.DataFrame(data=iris.target, columns=['target'])
target_df.tail(3)
```

```python
#target에 조건에 맞게 값 넣기
target_df.loc[target_df['target'] == 0, 'target'] = 'setosa'
target_df.loc[target_df['target'] == 1, 'target'] = 'versicolor'
target_df.loc[target_df['target'] == 2, 'target'] = 'virginica'
target_df.head(3)
```

```python
target_df.tail(3)
```

```python
df = pd.concat([iris_df, target_df], axis=1)
df.head()
```

```python
#컬럼 명 수정
df.columns = ['sepal_length',	'sepal_width', 'petal_length', 'petal_width',	'target']
df.head()
```

```python
def corr(df):
  plt.figure(figsize=(8, 8))
  sns.heatmap(df.corr(), annot = True, cmap = 'RdYlBu')
  plt.show()
```

```python
#상관관계 분석
df.corr()
```

```python
df.corr()
```

```python
t_list = list(df['target'].unique())

for i in t_list:
  print("========================[", i, "]=========================" )
  corr(df.loc[df['target'] == i].iloc[:, 0:4]) #petal_width 까지만 추출
```

### 유의성 검정

```python
#p_value 검증
#귀무가설 : 상관관계가 없다
#대립가설 : 상관관계가 있다

setosa_df = df.loc[df['target'] == 'virginica']
stats.pearsonr(x = setosa_df.petal_length, y = setosa_df.sepal_length) 

#x= 상관계수, y pvalue=0.0000..00629(거의 0에 가까운 값)
#pearson 상관계수가 0.864로, 강한 양의 상관관계 있다고 할 수 있음

#pvalue가 0에 가까운 값으로 0.05보다 작은 값이기 때문에, 귀무가설을 기각하고 대립가설을 채택한다(상관관계가 없다)
#0.05 미만인 경우 평균간에 차이가 없다(상관계수가 0이다=상관관계가 없다)는 귀무가설을 기각하고, 유의한 차이가 있다는 결론을 내림

#상관계수의 값은 크지만 p-value가 유의미하지 않은 경우 => 결과가 정확하지 않다는 결론
```

### 코호트 분석

```python
df = pd.read_excel('C:/Users/sju02/Desktop/데이터 분석/wakexmango/data/cohort_data.xlsx')
df.head()
```

```python
df.info() #str 가 아닌 datetime 형식
```

```python
df['order_date'] = df['order_date'].astype(str)
df['charges'] = df['charges'].astype(int)
```

```python
df.info()
```

```python
df.head(3)
```

```python
#cohort는 보통 daily 보다는 weekly나 monthly로 작업함:연,월,일 => 연.월 데이터로 바꾸기
#strptime : 원래 str로 있었던 데이터를 읽어와서 파싱하는것
#strftime : 포맷을 바꾸는것
df['order_ym']  = df['order_date'].map(lambda x : datetime.strptime(x, '%Y-%m-%d').strftime('%Y-%m'))
```

```python
df.head()
```

```python
#각 아이디 별 최초구매 년 월 구하기
tmp_order_ym = pd.DataFrame(df.groupby('user_id').order_ym.min()).reset_index() #user_id 인덱스로 되어있어서 reset_index
tmp_order_ym.columns = ['user_id', 'first_order_ym']
tmp_order_ym.head()
```

```python
df = pd.merge(df, tmp_order_ym, on=['user_id'], how='left')
df.head()
```

```python
#각 첫구매 월-재구매 월(첫구매 포함) -> 재구매율을 보는것이 목적
df.groupby(['first_order_ym', 'order_ym']).user_id.nunique() #유니크한 갯수
```

```python
#주문한 연월 - 맨 처음 주문한 연 월
df['order_ym'].map(lambda x : datetime.strptime(x, '%Y-%m'))-df['first_order_ym'].map(lambda x : datetime.strptime(x, '%Y-%m'))
```

```python
#처음 주문한 연 월
df['first_order_ym'].map(lambda x : datetime.strptime(x, '%Y-%m'))
```

```python
#코호트분석(특정 기간동한 cohort 별로 집계)


df['month_diff'] = round((df['order_ym'].map(lambda x : datetime.strptime(x, '%Y-%m')) \
    - df['first_order_ym'].map(lambda x : datetime.strptime(x, '%Y-%m')))/ np.timedelta64(1, 'M'))\ 
        .astype(int) #몇개월인지 나누기, round- 소수점 반올림 ,int 타입으로 바꿔서 소수점 제거
```

```python
df.head()
```

```python
pd.pivot_table(df, columns='month_diff', index='first_order_ym', values='charges', aggfunc='sum')
```

```python
cohort_month = pd.pivot_table(df, columns='month_diff', index='first_order_ym', values='charges', aggfunc='sum').fillna(0).astype(int)
cohort_month
```

```python
plt.figure(figsize=(22, 6))
ax = sns.heatmap(cohort_month, annot=True, fmt='d', linewidths = 1, cmap='cividis')
```

결론 -구매 금액) 2010-1 월에 들어온 사람들은 한달 뒤에도 돈을 많이 썼다. 이탈을 안하고 잔류를 많이 한다
- 해당 월에 cohort를 더 자세히 분석
- 어떤 마케팅을 했는지, 


```python
#user_id로 구매한 사람 수 cohort
cohort_month_user = pd.pivot_table(df, columns='month_diff', index='first_order_ym',\
     values='user_id', aggfunc=lambda x: len(x.unique())).fillna(0).astype(int)
cohort_month_user
```

```python
plt.figure(figsize=(22, 6))
ax = sns.heatmap(cohort_month_user, annot=True, fmt='d', linewidths = 1, cmap='cividis')
```

결론 - 구매 수 ) 2009-11 에 구매자는 많이 생겼지만 그게 지속되지는 않았다(개선점)
2010-1 잘 된 케이스라 develope 에 참고하기

```python
#위의 두개를 서로 나누어 인당 결제금액 보지
asp_df = round(cohort_month/cohort_month_user).fillna(0).astype(int)
asp_df
```

```python
plt.figure(figsize=(22, 6))
ax = sns.heatmap(asp_df, annot=True, fmt='d', linewidths = 1, cmap='cividis')
#보통 세로로 많이 본다
#2009-11 : 유저는 많이 들어 왔는데 구매단가가 낮다
```
